```python

def extract_text_from_pdf(pdf_path, output_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    with open(output_path, "w", encoding="utf-8") as f:
        f.write(text)
    print(f"Text extracted from {pdf_path} and saved to {output_path}")
```

```python
name_l=book_names[1].split("\\")
filename=name_l[-2]+"_"+name_l[-1].replace(".pdf",'')
name_l
filename
```

```python
#Path of the Input pdf
for i in range(0,len(book_names)):
       name_l=book_names[i].split("\\")
       filename=name_l[-2]+"_"+name_l[-1].replace(".pdf",'')
       extract_text_from_pdf(book_names[i],path+f'/CBSE_TextBook_txt/{filename}.txt')
#PDF_file = Path(book_names[0])
#PDF_file
```
